```python
import pandas as pd
vidname = "000001"
frame = "01"
name = "/home/dimoy/Documents/crashing/CrashBest/C_{vidname}_{frame}.jpg"
table:pd.DataFrame = pd.read_csv("Crash_Table.csv")
```

```python
frame = pd.DataFrame() 
window = 2
for i in range(1,52-window):
    frame_table = table[["vidname"]+[f"frame_{j}" for j in range(i,i+window)]]
    for j in range(i,i+window):
        frame_table.rename(columns={f"frame_{j}": f"frame_{j-i}"},inplace=True)
    frame_table["frame_index"] = i
    frame = pd.concat([frame,frame_table],ignore_index=True)
```

```python
frame
```

```python
from PIL import Image
import threading
 


def print_cube(index):
    img = Image.new("RGB",(1280,720*window))
    for i in range(window):
        frame_str = str(frame["frame_index"][index]+i).zfill(2)
        vidname_str = str(frame["vidname"][index]).zfill(6)
        img.paste(Image.open(f"/home/dimoy/Documents/crashing/CrashBest/C_{vidname_str}_{frame_str}.jpg"),(0,720*i))
    

    if(frame[f"frame_{window-1}"][index]):
        to_save = "dataset2/Crashing/"
    else:
        to_save = "dataset2/NotCrashing/"
    img.save(to_save+str(index)+".jpg")

    if(index%100==0):
        print(str(index/len(frame))+"% of dataset covered")
handle = []
for index, row in frame.iterrows():
    handle.append(threading.Thread(target = print_cube,args = (index,)))
    handle[-1].start()


for i in handle:
    i.join()
    
```
