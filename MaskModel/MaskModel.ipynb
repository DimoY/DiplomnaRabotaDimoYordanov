```python
import tensorflow as tf
```

```python
batch_size = 32
img_height = 64
img_width = 64

data_dir = "/kaggle/input/face-mask-detection/Dataset"
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size
)

val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size
)
```

```python
class_names = train_ds.class_names
num_classes = len(class_names)
```

```python
"Classes are: "+ ", ".join([i.replace("_"," ") for i in class_names])
```

```python

inputs = tf.keras.layers.Input((64,64,3))
conv1 = tf.keras.layers.Conv2D(32,7)(inputs)
maxpool1 = tf.keras.layers.MaxPool2D(3)(conv1)
conv1 = tf.keras.layers.Conv2D(32,1)(maxpool1)
conv1 = tf.keras.layers.Conv2D(32,3)(conv1)
maxpool1 = tf.keras.layers.MaxPool2D(3)(conv1)
flt = tf.keras.layers.Flatten()(maxpool1)
encoding1 = tf.keras.layers.Dense(128)(flt)
outputs = tf.keras.layers.Dense(3,activation="softmax")(encoding1)
model = tf.keras.Model(inputs,outputs)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
epochs=4
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

```python
model.evaluate(val_ds)
```

```python
model.save("MaskModel-0.91%")
```
